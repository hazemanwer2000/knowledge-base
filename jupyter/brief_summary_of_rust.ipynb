Rust is a statically-typed, compiled language, with no garbage collection (GC).

Rust's safety priniciples include,
- The presence of (heavy) compile-time checks.
- The insertion of run-time checks, by the compiler.
    - Note that some checks are only present in `debug` build(s), while others are additionally present in `release` build(s).
    - Note that a Rust program *panics* (i.e., exits with an error) when a run-time check fails.


| Compile-time check |
| --- |
| Access of a variable, whose object was *moved* (into another) |

| Run-time check | `debug` | `release` |
| --- | --- | --- |
| Integer overflow | ✔ | ✘ |
| Out-of-bound array access | ✔ | ✔ |

### Tuples <a class="anchor" id="tuples"></a>

A tuple is a collection of different types.

```rust
fn main()
{
    let tuple: (u8, u16, u32) = (0xFF, 0xFFFF, 0xFFFFFFFF);
    let x = 1;

    println!("{:?}", tuple);
    println!("{:?}", tuple.1);
}
```

### Arrays <a class="anchor" id="arrays"></a>

An array is a fixed-size list of the same type.

```rust
fn main()
{
    let array: [u8; 3] = [0xA, 0xB, 0xC];

    println!("{:?}", array);
    println!("{:?}", array[1]);
    println!("Size: {:?}", array.len());
}
```

## Variables <a class="anchor" id="variables"></a>

Variables are immutable by default.

```rust
fn main()
{
    let x: u8 = 1;

    /* error: cannot assign twice to immutable variable `x` */
    x = 2;

    println!("{:?}", x);
}
```

To define a mutable variable, use the `mut` keyword.

```rust
fn main()
{
    let mut x: u8 = 1;

    x = 2;

    println!("{:?}", x);
}
```

*Shadowing* allows the re-use of variable identifiers.

Shadowing of variables is possible within the same scope. 

```rust
fn main()
{
    let x: u8 = 1;

    let x: u8 = 2;

    println!("{:?}", x);

    let x: u8 = x + 1;

    println!("{:?}", x);
}
```

## Constants <a class="anchor" id="constants"></a>

A constant is evaluated at compile-time.

```rust
fn main()
{
    const X: u8 = 1;
    const Y: u8 = X + 1;

    println!("{:?}", X + Y);
}
```

```rust
fn main()
{
    let x: u8 = 1;
    /* error: attempt to use a non-constant value in a constant */
    const Y: u8 = x + 1;

    println!("{:?}", x + Y);
}
```

## Functions <a class="anchor" id="functions"></a>

A scope may be evaluated as an expression (and not a statement) if it contains an expression at its end.

```rust
fn main()
{
    let x: u8 = {
        let a: u8 = 1;
        let b: u8 = 2;
        let three: u8 = 3;
        a + b + three
    };

    println!("{:?}", x);
}
```

Hence, to return an expression, `<EXPR>`, a function's body may either:
- Use a `return <EXPR>;` statement anywhere within.
- Place `EXPR` at its end (recommended).

```rust
fn main()
{
    let x: u8 = add(1, 2);

    println!("{:?}", x);
}

fn add(x: u8, y: u8) -> u8
{
    let three: u8 = 3;
    x + y + three
}
```

## Control Flow <a class="anchor" id="control-flow"></a>

### `if` <a class="anchor" id="if"></a>

An `if` may be evaluated as an expression if all its scope(s) contain an expression at the end.

```rust
fn main()
{
    println!("{:?}", scale(1));
    println!("{:?}", scale(2));
    println!("{:?}", scale(3));
}

fn scale(x: u8) -> u8
{
    if (x == 1) { 2 } else if (x == 2) { 4 } else { 0 }
}
```

### `loop` <a class="anchor" id="loop"></a>

A `loop` executes a scope indefinitely, until a `break` statement is hit.

A `loop` may be evaluated as an expression by using `break <EXPR>` statement(s).

```rust
fn main()
{
    let mut x: u8 = 1;
    let y = loop {
        println!("Looping...");
        if (x < 5)
        {
            x = x + 1;
            continue;
        }
        break x;
    };

    println!("{:?}", y);
}
```

A `loop` may be labelled, and referenced explicitly by (nested) `break` and `continue` statement(s).

```rust
fn main()
{
    'a: loop {
        println!("Outer loop...");
        'b: loop {
            println!("Inner loop...");
            break 'a;
        }
    };
}
```

### `while` <a class="anchor" id="while"></a>

A `while` is a conditional alternative to `loop`.

A `while` may not be evaluated as an expression.

```rust
fn main()
{
    let mut x: u8 = 1;

    while (x <= 5)
    {
        println!("{:?}", x);
        x = x + 1;
    }
}
```

### `for` <a class="anchor" id="for"></a>

A `for` may be used to iterate through a iterable (e.g., an array).

A `for` may not be evaulated as an expression.

```rust
fn main()
{
    let slice: [u8; 3] = [0xA, 0xB, 0xC];

    for value in slice
    {
        println!("{:x}", value);
    }
}
```
